```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression, LogisticRegression
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

# Chọn các biến độc lập
X = data[['Age', 'Education', 'MaritalStatus', 'Occupation', 'Hours_spent_daily_online', 
          'Follow_favorite_brands_online', 'Interaction_with_brands_over_internet', 
          'Monthly_expenditure', 'Payment_method', 'FBattitude', 'CustPI', 
          'InteractionVar1', 'InteractionVar2', 'InteractionVar3', 
          'PCB', 'SNorms']]

# Biến phụ thuộc cho hồi quy tuyến tính
y_linear = data['Income']

# Biến phụ thuộc cho hồi quy logistic (chuyển đổi 'Income' thành nhị phân dựa trên trung vị)
y_logistic = (data['Income'] > data['Income'].median()).astype(int)

# Hồi quy tuyến tính
linear_model = LinearRegression()
linear_model.fit(X, y_linear)

# Tính toán R² và adjusted R² cho mô hình hồi quy tuyến tính
r_squared = linear_model.score(X, y_linear)
n = len(y_linear)
p = X.shape[1]
adjusted_r_squared = 1 - (1-r_squared) * (n-1)/(n-p-1)

# Hồi quy logistic
X_train, X_test, y_train, y_test = train_test_split(X, y_logistic, test_size=0.3, random_state=42)
logistic_model = LogisticRegression(max_iter=1000)
logistic_model.fit(X_train, y_train)

# Dự đoán và đánh giá hồi quy logistic
y_pred_logistic = logistic_model.predict(X_test)
accuracy_logistic = accuracy_score(y_test, y_pred_logistic)
conf_matrix_logistic = confusion_matrix(y_test, y_pred_logistic)
class_report_logistic = classification_report(y_test, y_pred_logistic)

# Kết quả hồi quy tuyến tính
coefficients_linear = linear_model.coef_
intercept_linear = linear_model.intercept_

# Kết quả hồi quy logistic
coefficients_logistic = logistic_model.coef_
intercept_logistic = logistic_model.intercept_

# In kết quả
print("Hồi quy tuyến tính:")
print(f"R²: {r_squared}")
print(f"Adjusted R²: {adjusted_r_squared}")
print(f"Hệ số hồi quy: {coefficients_linear}")
print(f"Hệ số chặn: {intercept_linear}")

print("\nHồi quy logistic:")
print(f"Độ chính xác: {accuracy_logistic}")
print("Ma trận nhầm lẫn:")
print(conf_matrix_logistic)
print("Báo cáo phân loại:")
print(class_report_logistic)
print(f"Hệ số hồi quy: {coefficients_logistic}")
print(f"Hệ số chặn: {intercept_logistic}")
```

```python
from sklearn.cluster import KMeans, AgglomerativeClustering, DBSCAN
from sklearn.metrics import silhouette_score
import matplotlib.pyplot as plt

# Sử dụng các biến được chọn để thực hiện phân tích cụm
X_clustering = data[['Age', 'Education', 'Hours_spent_daily_online', 
                     'Follow_favorite_brands_online', 'Interaction_with_brands_over_internet', 
                     'Monthly_expenditure']]

# Áp dụng phương pháp Elbow để xác định số lượng cụm tối ưu cho K-Means
sse = []
k_values = range(2, 11)  # Thử nghiệm với số cụm từ 2 đến 10

for k in k_values:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(X_clustering)
    sse.append(kmeans.inertia_)

# Vẽ biểu đồ Elbow
plt.figure(figsize=(10, 6))
plt.plot(k_values, sse, marker='o')
plt.title('Elbow Method for Optimal k')
plt.xlabel('Number of clusters (k)')
plt.ylabel('Sum of squared distances (Inertia)')
plt.show()
```

```python
# Áp dụng K-Means với số lượng cụm k=4 dựa trên Elbow Method
kmeans_final = KMeans(n_clusters=4, random_state=42)
clusters_final = kmeans_final.fit_predict(X_clustering)

# Tính toán Silhouette Score cho kết quả phân cụm K-Means
silhouette_avg_final = silhouette_score(X_clustering, clusters_final)

# Thêm kết quả phân cụm vào DataFrame gốc để phân tích đặc điểm từng cụm
data['Cluster_KMeans'] = clusters_final
cluster_summary_final = data.groupby('Cluster_KMeans').mean()

silhouette_avg_final, cluster_summary_final
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.decomposition import PCA

# Thực hiện phân tích thành phần chính (PCA) để giảm số chiều xuống 2D cho trực quan hóa
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_clustering)

# Thêm nhãn phân cụm vào dữ liệu PCA
data['PCA1'] = X_pca[:, 0]
data['PCA2'] = X_pca[:, 1]

# Vẽ biểu đồ phân cụm
plt.figure(figsize=(10, 8))
sns.scatterplot(x='PCA1', y='PCA2', hue='Cluster_KMeans', data=data, palette='viridis', s=100, alpha=0.7)
plt.title(f'PCA Plot of K-Means Clusters (k=4) with Silhouette Score: {silhouette_avg_final:.2f}')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.legend(title='Cluster')
plt.show()

# Hiển thị heatmap của các cụm
plt.figure(figsize=(12, 8))
sns.heatmap(cluster_summary_final.T, annot=True, fmt=".2f", cmap='coolwarm', linewidths=.5)
plt.title('Cluster Summary Heatmap')
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Tính toán tần suất mua sắm dựa trên 'Monthly_expenditure'
data['Purchase_Frequency'] = data.groupby('Monthly_expenditure')['Monthly_expenditure'].transform('count')

# Tạo bảng pivot để hiển thị mối quan hệ giữa Monthly_expenditure và các biến khác
heatmap_data = data.pivot_table(values='Purchase_Frequency',
                                index='Monthly_expenditure',
                                columns='Payment_method',
                                aggfunc='sum')

# Vẽ heatmap của tần suất mua sắm
plt.figure(figsize=(12, 8))
sns.heatmap(heatmap_data, annot=True, fmt='g', cmap='viridis')
plt.title('Heatmap of Purchase Frequency by Monthly Expenditure and Payment Method')
plt.xlabel('Payment Method')
plt.ylabel('Monthly Expenditure')
plt.show()

# Tạo heatmap cho mối quan hệ giữa Interaction_with_brands_over_internet và Follow_favorite_brands_online
heatmap_data2 = data.pivot_table(values='Purchase_Frequency',
                                 index='Interaction_with_brands_over_internet',
                                 columns='Follow_favorite_brands_online',
                                 aggfunc='sum')

# Vẽ heatmap của tần suất mua sắm dựa trên tương tác với thương hiệu và theo dõi thương hiệu yêu thích
plt.figure(figsize=(12, 8))
sns.heatmap(heatmap_data2, annot=True, fmt='g', cmap='viridis')
plt.title('Heatmap of Purchase Frequency by Interaction with Brands Online and Following Favorite Brands Online')
plt.xlabel('Follow Favorite Brands Online')
plt.ylabel('Interaction with Brands Online')
plt.show()
```
